```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Загрузка данных из CSV файла
file_path = r"results.csv"
data = pd.read_csv(file_path)

# Удаление лишних пробелов в названиях столбцов
data.columns = data.columns.str.strip()
```

```python
# Построение графиков
plt.figure(figsize=(12, 6))

# График времени выполнения на CPU (последовательное и многопоточное) и GPU
plt.subplot(1, 2, 1)
# Задать логарифмическую шкалу
plt.semilogy()
plt.plot(data['Matrix size'], data['CPU time (s)'], label='CPU', marker='o')
plt.plot(data['Matrix size'], data['CPU JIT time(s)'], label='CPU JIT', marker='o')
plt.plot(data['Matrix size'], data['CUDA time(s)'], label='CUDA', marker='o')
plt.plot(data['Matrix size'], data['Numpy dot time(s)'], label='Numpy dot', marker='o')
plt.title('Время выполнения на CPU и GPU')
plt.xlabel('Размер матрицы')
plt.ylabel('Время (секунды)')
plt.legend()
plt.grid()

# График ускорения
plt.subplot(1, 2, 2)
plt.plot(data['Matrix size'], data['CPU JIT speed up'], label='CPU JIT speed up', marker='o')
plt.plot(data['Matrix size'], data['CUDA speed up'], label='CUDA speed up', marker='o')
plt.plot(data['Matrix size'], data['Numpy dot speed up'], label='Numpy dot speed up', marker='o')
plt.title('Ускорение CPU и GPU')
plt.xlabel('Размер матрицы')
plt.ylabel('Ускорение')
plt.legend()
plt.grid()

# Показать графики
plt.tight_layout()
plt.show()
```
